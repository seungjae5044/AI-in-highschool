```
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_diabetes

import pandas as pd
#from sympy import symbols,

diabetes = load_diabetes()
print(diabetes.data.shape, diabetes.target.shape)

df_diabetesData=pd.DataFrame(diabetes.data,columns=['A','B','C','D','E','F','G','H','I','J'],dtype=float)
df_diabetesTarget=pd.DataFrame(diabetes.target)

print(df_diabetesData)
print(df_diabetesTarget)

plt.scatter(df_diabetesData['C'],df_diabetesTarget)
plt.show()
```

```
#define Linear_Regression:
class Neuron:
   
  def __init__(self):
    self.w=1.0
    self.b=1.0
    
  def forpass(self,x):
    y_hat=x*self.w+self.b
    return y_hat
  
  def backprop(self,x,err):
    w_grad=x*err
    b_grad=1*err
    return w_grad,b_grad
  
  def fit(self,x,y,epochs=100):
    for i in range(epochs):
      for x_i,y_i in zip(x,y):
        y_hat=self.forpass(x_i)
        err= -(y_i-y_hat)
        w_grad,b_grad=self.backprop(x_i,err)
        self.w=self.w-w_grad
        self.b=self.b-b_grad

x=df_diabetesData['C']
y=df_diabetesTarget[0]

neuron = Neuron()
neuron.fit(x,y)

plt.scatter(x,y)
print(neuron.w,neuron.b)
pt1= (-0.1,-0.1*neuron.w+neuron.b)
pt2= (0.15,0.15*neuron.w+neuron.b)
plt.plot([pt1[0],pt2[0]],[pt1[1],pt2[1]],color='green')
plt.show()
```
